**BOWLING SUMMARY**

```python
with open("bowling_summary.json") as f:
    data = json.load(f)
data
```

```python
df_bowling = pd.DataFrame(data)
df_bowling.head()
```

```python
with open("bowling_summary_updated.json") as f:
    data = json.load(f)
data
```

```python
df_bowling_up = pd.DataFrame(data)
df_bowling_up.head()
```

```python
merged_df = pd.merge(df_bowling_up, df_bowling, on=['Bowler_Name', 'Overs', 'Maidens', 'Runs', 'Wickets'], how='left')

# Step 3: Update Table 1
columns_to_update = ["Match_Between"]
df_bowling_up[columns_to_update] = merged_df[columns_to_update]
```

```python
df_bowling_up.head()
```

```python
df_bowling_up['match_id'] = df_bowling_up['Match_Between'].map(match_ids_dict)
df_bowling_up.head()
```

```python
df_bowling_up.to_csv('bowling_summary_updated.csv', index = False)
```

```python
df_bowling_up.to_json('bowling_summary_updated.json', orient='records')
```
